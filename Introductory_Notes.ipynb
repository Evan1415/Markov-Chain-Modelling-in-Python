### Classes 
Classes are a fundamental concept in Object Oriented Programming (OOP). They provide an extra layer of abstraction for writing programs in terms of "objects" on top of simply writing a program in terms of variables and functions. 

For instance, consider writing a program that calculates different properties belonging to a rectangle, e.g it's area. It's completely fine to write an area function that would take the length ($x$) and width ($y$) as arguments and return $A=x \times y$. But what if you also wanted to calculate the perimeter? It would be a lot easier to create a Rectangle object and extract from it important properties like area and perimeter. Chances are we will also be interested in more than one Rectangle and so being able to have a general ** blueprint ** of the object Rectangle would be very useful. 

In its essence, that is all a class is: a blueprint for the object that you want to create. When we create an object from using this blueprint we say we create an instance of the class itself.

This might be all very abstract, so consider the class for a rectangular shape: 

```python
#An example of a class
class Shape:
        
    def __init__(self,x,y):
        self.x = x
        self.y = y
    
    # method for area of the shape
    def area(self):
        return self.x * self.y
    
    # method for perimeter of the shape
    def perimeter(self):
        return 2 * self.x + 2 * self.y
    
    # method for the shape Shape class description
    def describe(self,text):
        self.description = text
        
    # method for storing the author of the Shape class
    def authorName(self,text):
        self.author = text
        
    # method for scaling the size of the shape
    def scaleSize(self,scale):
        self.x = self.x * scale
        self.y = self.y * scale
```

What we have above is a detailed description of a rectangular shape and. When an object is created it calls the __init__ function which creates the instance of the object. All off the methods (i.e functions) in the class can then be called by that instance object. Self is the way in which the object internally refers to itself. 

Let's create an instance of the above "Shape" class which can describe a rectangle.

```python
rectangle = Shape(10,10) # creates shape with x=10 and y=10
```

We can now extract the following information from our shape "Rectangle":

```python
area = rectangle.area()
perimeter = rectangle.perimeter()
rectangle.describe("Rectangle")
rectangle.authorName("Evan")
scaled = rectangle.scaleSize(0.1)
area2= rectangle.area()

print "The area of the shape is:", area
print "The permiter of the shape is:", perimeter
print "The shape is described as a ", rectangle.description
print "The author of the shape class is", rectangle.author
print "After re-scaling" , rectangle.x, area2
```

Numpy **Arrays** allow for the creation of vectors and matrices that can be efficiently manipulated with the internal numpy routines. They are *statically typed* which implies an upper hand on doing such operations on lists for example. 

```python
import numpy as np

vector1 = np.array([1,2])

matrix1 = np.array([[1,2],[1,3]])

print "Vector 1 is:", vector1, "with dimensions", np.shape(vector1)
print "Matrix 1 is:", matrix1, "with dimensions", np.shape(matrix1)
```

For larger arrays, it isn't feasible to create them by hand. There a number of different `numpy` functions that can do this automatically. For example:

```python
start = 0  
stop = 10 
step = 1

x = np.arange(start, stop, step) # generates array from start -> stop in increments of step\
print x

# to include the last point, i.e 10, we'd have to use the numpy function "linspace"

step_num = np.abs((stop-start) + 1)/step # number of points in the interval

x = np.linspace(start,stop,step_num) # array now spans from stop -> start inclusively
print x
```

We can also generate random arrays in numpy by using `np.random` 

```python
# uniform random numbers in [0,1]
x = np.random.rand(5,5)
print x
```

`Numpy` has a number of mathematical functions that can be used on arrays. For instance, let's say we want to take the cosine of the above random array, all we have to do is:

```python
y = np.cos(x)
print y
```

Finally, numpy can read in data files (it is not the most efficient way to do this in general, see `Pandas`). To do so, we will first save an array to a file and then read it back in:

## Plotting in Python 

One of my favourite things about python is its extremely versatile and easy to use plotting capabilities. In this tutorial we are going to learn how to use Matplotlib: a Python library for plotting. 

To get started with Matplotlib, we first need to import the library and common practice is to relabel it "plt" for short

```python
%matplotlib inline 
# the above allows for plots to be inset in this IPython Notebook
import numpy as np
import matplotlib # imports the entire namespace of functions belonging to matplotlib
import matplotlib.pyplot as plt # shortens matplotlib.pyplot -> plt
```

Matplotlib is a nice illustration of the power of object oriented programming. The library has a figure class on which methods are defined so we can manipulate the figures. 

For example, we can store a figure in the variable "fig":

```python
fig = plt.figure()
x = np.linspace(0, 10, 100) # creates array from 0->100 with 100 elements in increments of 10/100
y = np.sin(x) # evaluates sin(x) on the array x
axes = fig.add_axes([0,0,1,1]) # left, bottom, width, height: defines size of image
axes.plot(x, y, 'r') # plots y(x), 'r': colour red
axes.set_xlabel('x-axis label')
axes.set_ylabel('y-axis label')
axes.set_title('My first plot');
```

The fun doesn't stop here, the Matplotlib figure object has many other methods that have the potential to make the above plot a lot more detailed and impressive. For example, it is straightforward to generate a graph with a smaller one inside of it. 

```python
x2 = np.linspace(0,10,100)
y2 = x2**2

fig2 = plt.figure()

axes1 = fig2.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
axes2 = fig2.add_axes([0.2, 0.5, 0.4, 0.3]) # inset axes and position of graph 

axes1.grid(color='b', linestyle='dashed', linewidth=.5)
axes2.grid(False)

# main figure
axes1.plot(x2, y2, 'r')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('Main title')

# inset
axes2.plot(x2, y2, 'g')
axes2.set_xlim([0,1])
axes2.set_ylim([0,1])
axes2.set_title('Inset title');

plt.savefig("inset.pdf") # saves the image to file
```

It is also possible to include a legend with a specific font and LaTeX syntax:

```python
fig3 = plt.figure()
axes3 = fig3.add_axes([0.1, 0.1, 1, 1])
matplotlib.rcParams.update({'font.size': 18, 'font.family': 'serif'})
axes3.plot(x2,y2,'r',label = r"$y = \alpha^2$")
axes3.set_xlabel(r"$\alpha$")
axes3.set_ylabel(r"$\alpha^2$")
axes3.legend(loc=2)
```

It is useful to know how to do 3d plots and contours. Below is an example (albeit a little complicate) of such a plot:

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
alpha = 0.7
phi_ext = 2 * np.pi * 0.5

def flux_qubit_potential(phi_m, phi_p):
    return 2 + alpha - 2 * np.cos(phi_p) * np.cos(phi_m) - alpha * np.cos(phi_ext - 2*phi_p)

phi_m = np.linspace(0, 2*np.pi, 100)
phi_p = np.linspace(0, 2*np.pi, 100)
X,Y = np.meshgrid(phi_p, phi_m)
Z = flux_qubit_potential(X, Y).T
fig = plt.figure(figsize=(8,6))


fig, ax = plt.subplots()

cnt = ax.contour(Z, cmap=matplotlib.cm.RdBu, vmin=abs(Z).min(), vmax=abs(Z).max(), extent=[0, 1, 0, 1])
```

```python
fig = plt.figure(figsize=(8,6))

ax = fig.add_subplot(1,1,1, projection='3d')

ax.plot_surface(X, Y, Z, rstride=4, cstride=4, alpha=0.25)
cset = ax.contour(X, Y, Z, zdir='z', offset=-np.pi, cmap=matplotlib.cm.coolwarm)
cset = ax.contour(X, Y, Z, zdir='x', offset=-np.pi, cmap=matplotlib.cm.coolwarm)
cset = ax.contour(X, Y, Z, zdir='y', offset=3*np.pi, cmap=matplotlib.cm.coolwarm)

ax.set_xlim3d(-np.pi, 2*np.pi);
ax.set_ylim3d(0, 3*np.pi);
ax.set_zlim3d(-np.pi, 2*np.pi);

plt.savefig("example.pdf")
```
